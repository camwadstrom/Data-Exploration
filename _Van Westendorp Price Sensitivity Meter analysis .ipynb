```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data

#q1: "At what price would you consider the product to be too expensive?"
q1 = [90, 60, 75, 90, 90, 105, 110, 120, 110, 140]

#q2: "At what price point would you consider the product to be a bargain?"
q2 = [50, 40, 40, 50, 60, 70, 85, 85, 90, 90]

#q3: "At what price would you consider the product to be getting expensive, but you would still consider purchasing?"
q3 = [60, 65, 80, 85, 95, 90, 100, 100, 75, 100]

#q4: "At what price would you consider the product to be at a good value?"
q4 = [50, 55, 60, 70, 85, 85, 90, 90, 75, 60]


# Function to calculate cumulative percentage
def calculate_cumulative_percentage(data):
    data.sort()
    total_responses = len(data)
    cumulative_percentage = [(i + 1) / total_responses * 100 for i in range(total_responses)]
    return data, cumulative_percentage

# Calculate cumulative percentages for each question
q1_sorted, q1_cumulative = calculate_cumulative_percentage(q1)
q2_sorted, q2_cumulative = calculate_cumulative_percentage(q2)
q3_sorted, q3_cumulative = calculate_cumulative_percentage(q3)
q4_sorted, q4_cumulative = calculate_cumulative_percentage(q4)

# Plotting the cumulative percentage distribution curves
plt.figure(figsize=(8, 6))
plt.plot(q1_sorted, q1_cumulative, label='Too Expensive')
plt.plot(q2_sorted, q2_cumulative, label='Bargain')
plt.plot(q3_sorted, q3_cumulative, label='Getting Expensive')
plt.plot(q4_sorted, q4_cumulative, label='Good Value')

# Highlight intersecting or overlapping points
intersecting_values = list(set(q1_sorted) & set(q2_sorted) & set(q3_sorted) & set(q4_sorted))
intersecting_percentages = [(i + 1) / len(intersecting_values) * 100 for i in range(len(intersecting_values))]
plt.scatter(intersecting_values, intersecting_percentages, color='red', label='Intersecting Points', marker='o')

plt.title('Van Westendorp Price Sensitivity Meter')
plt.xlabel('Price')
plt.ylabel('Cumulative Percentage')
plt.legend()
plt.grid(True)
plt.show()

# Display intersecting values and their cumulative percentages
print("Intersecting Values:", intersecting_values)
print("Intersecting Cumulative Percentages:", intersecting_percentages)
```
